```python
def get_data_info(data_info_path):
    data_info = defaultdict(int)
    with open(data_info_path, "r") as f:
        for ln in f:
            if not ln:
                break
            record = json.loads(ln)
            if isinstance(record, str):
                # This means record is still a JSON string, so decode again
                record = json.loads(record)
            # Grab the date in YYYY-MM-DD format
            date_str = record["tweet_date"][:10]
            
            # Here we assume there's a key named "count" or similar
            # that tells us how many total items are relevant for that date.
            data_info[date_str] += 1
    return data_info
```

```python
def plot_class_percentages(start_date, end_date, by_date, data_info, plt_name = "perc.png"):
    from datetime import datetime, timedelta
    import matplotlib.pyplot as plt

    # Convert start/end to datetime objects
    start_date_obj = datetime.strptime(start_date, "%Y-%m-%d")
    end_date_obj = datetime.strptime(end_date, "%Y-%m-%d")

    # Build a list of all dates in [start_date, end_date]
    date_range = [
        start_date_obj + timedelta(days=i)
        for i in range((end_date_obj - start_date_obj).days + 1)
    ]

    # Prepare lists for Class 1% and Class 2%
    class_1_pct = []
    class_2_pct = []
    x_dates = []  # for the x-axis
    
    for d in date_range:
        day_str = d.strftime("%Y-%m-%d")
        total_for_day = data_info.get(day_str, 0)
        c1_count = by_date[1].get(day_str, 0)
        c2_count = by_date[2].get(day_str, 0)

        if total_for_day == 0:
            # Avoid dividing by zero. Could set to 0% or skip the date
            c1_percent = 0.0
            c2_percent = 0.0
        else:
            c1_percent = (c1_count / total_for_day) * 100.0
            c2_percent = (c2_count / total_for_day) * 100.0

        class_1_pct.append(c1_percent)
        class_2_pct.append(c2_percent)
        x_dates.append(d)

    # Now plot the two lines
    plt.figure(figsize=(10, 6))
    plt.plot(x_dates, class_1_pct, 'r', label="Class 1 (%)")
    plt.plot(x_dates, class_2_pct, 'g', label="Class 2 (%)")

    plt.title(f"Class 1 & Class 2 Tweets (as % of All) from {start_date} to {end_date}")
    plt.xlabel("Date")
    plt.ylabel("Percentage of All Tweets")
    plt.legend()
    plt.grid(True)
    plt.savefig(plt_name)
    plt.show()
```

```python
def plot_range(start_date, end_date, data_info, include_class_3=False, stackplot=True):
    """
    start_date, end_date: 'YYYY-MM-DD' strings
    data_info: a dict[date_str] -> total items for that day (from the second file)
    include_class_3: whether to include class 3 in the stack plot
    stackplot: if True, plot proportions in a stack plot; if False, plot raw counts
    """
    from datetime import datetime, timedelta
    import matplotlib.pyplot as plt

    # Convert user-supplied strings to datetime objects
    start_date_obj = datetime.strptime(start_date, "%Y-%m-%d")
    end_date_obj = datetime.strptime(end_date, "%Y-%m-%d")

    # Make a list of all days in [start_date, end_date]
    date_range = [
        start_date_obj + timedelta(days=i)
        for i in range((end_date_obj - start_date_obj).days + 1)
    ]

    # Get raw counts from your by_date dictionary
    class_1_counts = [by_date[1][d.strftime("%Y-%m-%d")] for d in date_range]
    class_2_counts = [by_date[2][d.strftime("%Y-%m-%d")] for d in date_range]
    class_3_counts = [by_date[3][d.strftime("%Y-%m-%d")] for d in date_range]

    if stackplot:
        # When stackplot=True, we will plot normalized proportions:
        # each class's count / sum_of_counts  (within the labeled data)
        # or you might want to do: class_count / data_info[day_str] if your
        # second file represents the true total. Adapt as needed.

        # A. Using only the labeled data counts for normalization:
        sums = [
            (sum(x) if sum(x) > 0 else 1e-7)
            for x in (zip(class_1_counts, class_2_counts, class_3_counts) if include_class_3
                      else zip(class_1_counts, class_2_counts))
        ]
        c1_norm = [c / s for c, s in zip(class_1_counts, sums)]
        c2_norm = [c / s for c, s in zip(class_2_counts, sums)]
        c3_norm = [c / s for c, s in zip(class_3_counts, sums)]

        plt.figure(figsize=(10, 6))
        if include_class_3:
            plt.stackplot(date_range, c1_norm, c2_norm, c3_norm,
                          colors=['r', 'g', 'b'], labels=["Class 1", "Class 2", "Class 3"])
        else:
            plt.stackplot(date_range, c1_norm, c2_norm,
                          colors=['r', 'g'], labels=["Class 1", "Class 2"])
        plt.ylabel("Proportion of Tweets (Based on Labeled Data)")
        plt.title("Proportion of Tweets from " + start_date + " to " + end_date)

    else:
        # If stackplot=False, we will plot *absolute counts* or *percentages from data_info*.
        # Suppose we want to use data_info to compute a ratio for class_1 and class_2.
        # For example: class_1_ratio = class_1_counts[i] / data_info[day_str]
        # Then we plot the ratio instead of raw counts.

        c1_ratio = []
        c2_ratio = []
        c3_ratio = []

        for i, d in enumerate(date_range):
            day_str = d.strftime("%Y-%m-%d")
            # Get total from the second dictionary
            total_for_day = data_info.get(day_str, 0)
            total_for_day = total_for_day if total_for_day > 0 else 1e-7  # avoid div by zero

            c1_ratio.append(class_1_counts[i] / total_for_day)
            c2_ratio.append(class_2_counts[i] / total_for_day)
            c3_ratio.append(class_3_counts[i] / total_for_day)

        # Now plot these as lines (these represent percentages or fractions)
        plt.figure(figsize=(10, 6))
        plt.plot(date_range, c1_ratio, 'r', label="Class 1 %")
        plt.plot(date_range, c2_ratio, 'g', label="Class 2 %")
        if include_class_3:
            plt.plot(date_range, c3_ratio, 'b', label="Class 3 %")

        plt.ylabel("Percentage (Class Count / data_info[day])")
        plt.title("Percentage from " + start_date + " to " + end_date)

    plt.xlabel("Date")
    plt.legend()
    plt.show()
```

```python
import json
import pandas as pd

# Path to your JSON file
json_file_path = "./annotatedv2_70b_prompt1_v5.json"

# Initialize an empty list to collect processed chunks
chunk_size = 10000  # Number of lines to process at a time
chunks = []

# Read the file line by line
with open(json_file_path, "r", encoding="utf-8") as file:
    chunk = []
    for i, line in enumerate(file):
        try:
            chunk.append(json.loads(line.strip()))
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON at line {i}: {e}")
        
        # Process the chunk when it reaches the specified size
        if (i + 1) % chunk_size == 0:
            df_chunk = pd.json_normalize(chunk)
            chunks.append(df_chunk)
            chunk = []  # Clear the chunk to save memory
            print(f"Processed {i + 1} lines...")

    # Process any remaining lines
    if chunk:
        df_chunk = pd.json_normalize(chunk)
        chunks.append(df_chunk)

# Combine all chunks into a single DataFrame
df = pd.concat(chunks, ignore_index=True)

# Save to a CSV file (optional)
df.to_csv("processed_data.csv", index=False)

print("Processing complete!")
```

```python
# Updated function to generate word clouds with bigrams
from wordcloud import WordCloud

def generate_word_clouds_with_bigrams(df):
    for cls in [1, 2]:  # Focus on Classes 1 and 2
        # Filter tweets for the current class
        class_text = df[df["classification"] == cls]["tweet_text"]
        
        # Combine all text for the class
        text = " ".join(class_text)
        
        # Generate bigrams using CountVectorizer
        vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words=[
            "https", "mi", "bunu", "ise", "hem", "size", "sadece", "sonra", "imdi", "zaman", "m",
            "demek", "ey", "artk", "var", "ben", "diye", "nasl", "kim", "ok", "kadar", "senin",
            "sizin", "bile", "seni", "hi", "bak", "yine", "asla", "bu", "de", "ama", "bir", "ne",
            "yok", "co", "RT", "t", "ve", "da", "o", "zaten", "ile", "siz", "sen", "ya", "ki",
            "deil", "neden", "gibi"
        ])
        bigram_matrix = vectorizer.fit_transform([text])
        bigram_counts = dict(zip(vectorizer.get_feature_names_out(), bigram_matrix.toarray().sum(axis=0)))

        # Generate the word cloud
        wordcloud = WordCloud(
            width=800,
            height=400,
            background_color="white",
            colormap="viridis",
        ).generate_from_frequencies(bigram_counts)

        # Plot the word cloud
        plt.figure(figsize=(10, 6))
        plt.imshow(wordcloud, interpolation="bilinear")
        plt.axis("off")
        plt.title(f"Word Cloud for Class {cls} (Bigrams)")
        plt.savefig(f"figures/word_cloud_class_{cls}_bigrams.png")
        plt.show()
generate_word_clouds_with_bigrams(df)
```

```python
# Filter classes with more than 20 samples
valid_classes = df["classification"].value_counts()[df["classification"].value_counts() > 20].index
filtered_df = df[df["classification"].isin(valid_classes)]

# Normalize per class
def normalize_per_class(df, group_col, normalize_col):
    """Normalizes counts within each class."""
    counts = df.groupby([group_col, normalize_col]).size().reset_index(name="count")
    total_per_class = counts.groupby(group_col)["count"].transform("sum")
    counts["proportion"] = counts["count"] / total_per_class
    return counts

# Gender normalization
gender_counts = normalize_per_class(filtered_df, "classification", "gender")

# Age group normalization
age_group_counts = normalize_per_class(filtered_df, "classification", "age_group")

# Plot Gender Distribution (Normalized)
def plot_normalized_distribution(counts, x_col, hue_col, title, filename):
    plt.figure(figsize=(12, 6))
    sns.barplot(
        x=x_col,
        y="proportion",
        hue=hue_col,
        data=counts,
        palette="viridis",
    )
    plt.title(title)
    plt.ylabel("Proportion")
    plt.xlabel("Class")
    plt.savefig(filename)
    plt.show()

# Plot Gender Distribution
plot_normalized_distribution(
    gender_counts,
    x_col="classification",
    hue_col="gender",
    title="Normalized Gender Distribution by Class",
    filename="figures/normalized_gender_distribution_by_class.png",
)

# Plot Age Group Distribution
plot_normalized_distribution(
    age_group_counts,
    x_col="classification",
    hue_col="age_group",
    title="Normalized Age Group Distribution by Class",
    filename="figures/normalized_age_group_distribution_by_class.png",
)
```

```python
# Function to clean noisy data
def clean_emotion_values(value):
    if value == "True":
        return True
    elif value == "False" or pd.isna(value):
        return False
    elif value == True:
        return True
    else:
        return False  # Treat noisy or unexpected values as False

# Apply cleaning to all emotion columns
emotion_columns = [col for col in df.columns if col.startswith("emotions.")]
df[emotion_columns] = df[emotion_columns].applymap(clean_emotion_values)

# Filter emotion columns and classification
df_emotions = df[emotion_columns + ["classification"]]

# Group by class and calculate the sum of True values for each emotion
class_emotions = df_emotions.groupby("classification").sum()

# Normalize the emotion counts for each class
class_emotions_normalized = class_emotions.div(class_emotions.sum(axis=1), axis=0)

# Plot normalized emotion distribution
def plot_normalized_emotions(normalized_emotions):
    normalized_emotions.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="coolwarm")
    plt.title("Normalized Emotion Distribution by Class")
    plt.xlabel("Class")
    plt.ylabel("Proportion")
    plt.legend(title="Emotions", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.savefig("figures/normalized_emotion_distribution_by_class.png")
    plt.show()

plot_normalized_emotions(class_emotions_normalized)
```
